# **Performing univariate EDA on healthcare-dataset-stroke-data**

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

url = "https://raw.githubusercontent.com/karavokyrismichail/Stroke-Prediction---Random-Forest/main/healthcare-dataset-stroke-data/healthcare-dataset-stroke-data.csv"
df = pd.read_csv(url)

# Understaning data
print("Shape of data :- ")
print(df.shape)
print()

print("Having a look at some example of dataset.")
print(df.sample(5))
print()

print("Information about data types used in dataset")
print(df.info())
print()

print("Finding whether missing values are prenset in data or not ")
print(df.isnull().sum())
print()

print("Are there any duplicate values:- ")
print(df.duplicated().sum())
print()

print("Columns of the dataset :- ")
columns = df.columns
print(columns)
print()

numeric_columns = df.select_dtypes(include=['number'])
print(numeric_columns.corr())
```

**Performing Univariate EDA on  Numerical Data**

```python
for col in ['age' , 'hypertension' , 'heart_disease' , 'avg_glucose_level' , 'bmi' , 'stroke']:
  print(df[col].describe())
  print("Number of missing values : " , df[col].isnull().sum())
  print()
  print("Ploting histogram for all the numerical data :")
  sns.histplot(df[col] , kde=True)
  plt.title(col)
  plt.show()
  print()
  print("Ploting bar plot :")
  print()
```

# Univariate Analysis on Stroke Prediction Dataset

## **1. Age Distribution**
- The age distribution appears **uniform** if the histogram is evenly spread.
- If skewed, it means there are **more younger or older people** in the dataset.

## **2. Hypertension & Heart Disease**
- Majority of people **do not have hypertension or heart disease**, indicating a generally healthy population.
- However, this **does not confirm that they are not at risk of stroke**—further analysis is needed.

## **3. Average Glucose Level**
- If most values are **below 100 mg/dL**, it suggests normal glucose levels for the majority.
- Since **diabetes is a major risk factor for stroke**, we need to analyze glucose levels among stroke patients.

## **4. BMI (Body Mass Index)**
- The **average BMI is 28-30**, which falls in the **overweight (25-29.9) or mild obesity (>30) category**.
- Since obesity is a **known risk factor for stroke**, further investigation is required.

## **5. Majority of People Are Healthy & Did Not Get a Stroke**
- The dataset shows that most people did not have a stroke.
- However, this **does not mean the risk factors (hypertension, heart disease, high BMI) do not contribute to stroke occurrence**.
