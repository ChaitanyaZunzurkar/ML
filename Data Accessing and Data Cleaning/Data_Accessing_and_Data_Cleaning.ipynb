### Data Cleaning

```python
patients.fillna('No data' , inplace=True) # replace all missing values with No data
```

```python
patients['address'].isnull().sum()
```

```python
treatments_full['hba1c_change'] = treatment['hba1c_start'] - treatment['hba1c_end'] # fill all the missing values in hba1c_changes
```

```python
treatments_long = treatments_full.melt(
    id_vars=['given_name', 'surname', 'hba1c_start', 'hba1c_end', 'hba1c_change'],
    var_name='Type',
    value_name='Dosage'
)

# Now you can safely filter
treatments_long = treatments_long[treatments_long['Dosage'] != '-']
treatments_long
```

```python
treatments_long['dosage_start'] = treatments_long['Dosage'].str.split('-').str.get(0)
treatments_long['dosage_end'] = treatments_long['Dosage'].str.split('-').str.get(1)
```

```python
treatments_long
```

```python
treatments_long['dosage_start'] = treatments_long['dosage_start'].str.replace('u' , '')
treatments_long['dosage_end'] = treatments_long['dosage_end'].str.replace('u' , '')
```

```python
treatments_long
```

```python
treatments_long = treatments_long.merge(adverse_reactions , how='left' , on=['given_name' , 'surname'])
```

```python
treatments_long['adverse_reaction'].fillna('No Adverse Reaction' , inplace=True)
```
