```python
df1.sample(5)
```

```python
has_5G = df1['Sim'].str.contains('5G')
has_NFC = df1['Sim'].str.contains('NFC')
has_IR_Blaster = df1['Sim'].str.contains('IR Blaster')
```

```python
df1.insert(7 , 'has_5G' , has_5G )
df1.insert(8 , 'has_NFC' , has_NFC )
df1.insert(9 , 'has_IR_Blaster' , has_IR_Blaster )
```

```python
df1.drop(index=405 , inplace=True) # its not a smartphone
```

```python
df1['Processor'].value_counts()
```

```python
df1['Processor'].str.split(',' , expand=True)
```

```python
df1['Processor Name'] = df1['Processor'].str.split(',' , expand=True)[0]
df1['Number of Cors'] = df1['Processor'].str.split(',' , expand=True)[1]
df1['Processor Speed'] = df1['Processor'].str.split(',' , expand=True)[2]
```

```python
display(df1['Processor Name'].value_counts())
display(df1['Number of Cors'].value_counts())
df1['Processor Speed'].value_counts()
```

```python
df1.columns = df1.columns.str.strip()
```

```python
df1['Number of Cors'] = df1['Number of Cors'].str.replace('Octa Core Processor', 'Octa Core', regex=False)
df1['Number of Cors'] = df1['Number of Cors'].str.replace('Nine Core', 'Nine Cores', regex=False)
df1['Number of Cors'] = df1['Number of Cors'].str.replace('Nine-Cores', 'Nine Cores', regex=False)
df1['Number of Cors'] = df1['Number of Cors'].str.replace('Nine Coress', 'Nine Cores', regex=False)
```

```python
df1['Number of Cors'].value_counts()
```

```python
df1.rename(columns={'RAM': 'RAM and ROM'} , inplace=True)
```

```python
df1['RAM'] = df1['RAM and ROM'].str.split(',' , expand=True)[0]
df1['ROM'] = df1['RAM and ROM'].str.split(',' , expand=True)[1]
```

```python
df1['RAM'].value_counts()
```

```python
df1[df1['RAM'] == 'Unisoc T107']
```

```python
df1.drop(index=[443 , 682 , 768] , inplace=True)
```

```python
df1['RAM'].value_counts()
```

```python
condition = df1['RAM'].str.contains('inbuilt')
df1.loc[condition , ['RAM' , 'ROM']] = df1.loc[condition , ['ROM' , 'RAM']].values
```

```python
df1['RAM'].value_counts()
```

```python
df1['Battery'].value_counts()
```

```python
split_battery = df1['Battery'].str.split('with', expand=True)

df1['Battery Capacity'] = split_battery[0]
df1['Fast Charging'] = split_battery[1]
```

```python
df1['Fast Charging'].value_counts()
```

```python
df1['Battery Capacity'].value_counts()
```

```python
df1.drop('Sim' , axis=1 , inplace=True)
df1.drop('Processor' , axis=1 , inplace=True)
df1.drop('RAM and ROM' , axis=1 , inplace=True)
df1.drop('Battery' , axis=1 , inplace=True)
```

```python
df1['Display Size'].value_counts()
```

```python
df1[['Display size', 'Resolution', 'Display Info']] = df1['Display Size'].str.split(',', n=2, expand=True)

df1['Display size'] = df1['Display size'].str.strip()
df1['Resolution'] = df1['Resolution'].str.strip()
df1['Display Info'] = df1['Display Info'].str.strip()

df1['Refresh Rate'] = df1['Display Info'].str.extract(r'(\d+\s*Hz)', expand=False)

df1['Display Feature'] = df1['Display Info'].str.extract(r'(Punch Hole|Notch|Waterdrop|Infinity|Dot)', expand=False)
```

```python
df1['Display Info'].value_counts()
```

```python
df1.drop('Display Info' , axis=1 , inplace=True)
```

```python
df1.drop('Display Size' , axis=1 , inplace=True)
```

```python
df1['Camera'].value_counts()
```

```python
df1['Rear Camera'] = df1['Camera'].str.split('&' , expand=True)[0]
df1['Front Camera'] = df1['Camera'].str.split('&' , expand=True)[1]
```

```python
df1.drop('Camera' , axis=1 , inplace=True)
```

```python
df1['Foldable and Dual Display'].replace({'Not Foldable and Dual Display': 'No Foldable Display , No Dual Display'} , inplace=True)
```

```python
df1['Foldable and Dual Display'].value_counts()
```

```python
df1['Foldable Display'] = df1['Foldable and Dual Display'].str.split(',' , expand=True)[0]
df1['Dual Display'] = df1['Foldable and Dual Display'].str.split(',' , expand=True)[1]
```

```python
df1['Foldable Display'].value_counts()
```

```python
df1[df1['Foldable Display'] == 'Dual Display']
```

```python
conditions = df1['Foldable Display'] == 'Dual Display'
df1.loc[conditions , ['Foldable Display' , 'Dual Display']] = df1.loc[conditions , ['Dual Display' , 'Foldable Display']].values
```

```python
df1['Dual Display'] = df1['Dual Display'].str.strip()
```

```python
df1['Foldable Display'].value_counts()
```

```python
df1.drop('Foldable and Dual Display' , axis=1 , inplace=True)
```
